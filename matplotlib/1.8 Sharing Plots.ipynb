```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# import heart disease data
heart_disease = pd.read_csv('heart-disease.csv')

# limit data to patients over the age of 50
over_50 = heart_disease[heart_disease['age'] > 50]

# Print first 5 rows of data
over_50.head()
```

```python
# Recreate plots from previous exercise
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               sharex=True,
                               figsize=(10, 10))

# Plot ax0
scatter=ax0.scatter(over_50['age'],
                    over_50['chol'],
                    c=over_50['target'],
                    cmap='coolwarm')               

ax0.set(title='Heart Disease and Cholesterol Levels',
        ylabel='Cholesterol')

ax0.legend(*scatter.legend_elements(),
           title='Target')        

ax0.axhline(y=over_50['chol'].mean(),
            color='orange',
            linestyle='-.',
            label='Average');

# Plot ax1
scatter=ax1.scatter(over_50['age'],
                    over_50['thalach'],
                    c=over_50['target'],
                    cmap='spring')
                    
ax1.set(title='Heart Disease and Maximum Heart Rate Levels',
        xlabel='Age',
        ylabel='Max Heart Rate')

ax1.legend(*scatter.legend_elements(),
           title='Target')

ax1.axhline(y=over_50['thalach'].mean(),
            color='b',
            linestyle=':', # supported values are '-', '--', '-.', ':'
            label='Average');

# Title the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
